```python
import yfinance as yf
import mplfinance as mpf
import warnings
warnings.filterwarnings('ignore')

data = yf.download("AAPL", start="2022-01-01", end="2022-06-30")
```

```python
mpf.plot(data)#The default chart type is a simple OHLC plot.
```

```python
mpf.plot(data, type="candle")#adding the type argument, we can define the plot type.
```

```python
mpf.plot(data, type="line")#simple line charts
```

```python
mpf.plot(data, type="renko")#mpf.plot(data, type="renko")
```

```python
mpf.plot(data, type="ohlc", mav=15)
# include moving averages to the plot. 
# Use a single number for a single moving average.
# Use a tuple of numbers for several.
```

```python
mpf.plot(data, type="candle", mav=(7, 14, 21))
# include three moving averages
# The result is a candle chart with 7, 15, and 21 day moving averages
```

```python
mpf.plot(data, type="candle", mav=(7, 14, 21), volume=True)
```

```python
intraday = yf.download(tickers="AAPL", period="5d", interval="1m")
iday = intraday.iloc[-100:, :]
mpf.plot(iday, type="candle", mav=(7, 12), volume=True)
# 1-minute chart showing the last 100 minutes of prices.
```
